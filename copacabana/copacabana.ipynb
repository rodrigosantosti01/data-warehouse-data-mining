# Exercício 03. Construa gráficos adequados para representar:
b) Copacabana: preço X distância da praia; área construída X distância da praia

```python
plt.figure(figsize=(15, 5))
sorted_df = df.sort_values(by=['VAL_UNIT'], axis=0, ascending=False)
# sorted_df
```

```python
sns.barplot(x='DistPraia', y='VAL_UNIT', data=sorted_df[:20])
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sorted_df = df.sort_values(by=['AreaConstruida'], axis=0, ascending=False)
# sorted_df
```

```python
sns.barplot(x='AreaConstruida', y='DistPraia', data=sorted_df[:20])
plt.show()
```
